```python
# Problem 1: Post Engagement Analyzer
def analyze_engagement(likes):
    total_likes = sum(likes)
    status = "Viral Post" if total_likes >= 1000 else "Normal Engagement"
    print("Total Likes:", total_likes)
    print("Post Status:", status)

analyze_engagement([200, 300, 250, 300])
```

```python
# Problem 2: Medicine Stock Alert System

def medicine_stock(stock):
    status = "Low Stock Alert" if stock < 10 else "Stock Sufficient"
    print("Medicine Stock:", stock)
    print("Status:", status)

medicine_stock(6)
```

```python
# Problem 3: Rainfall Adequacy Checker
def rainfall_checker(rainfall, required_level):
    avg = sum(rainfall) / len(rainfall)
    status = "Adequate Rainfall" if avg >= required_level else "Inadequate Rainfall"
    print("Average Rainfall:", avg)
    print("Rainfall Status:", status)

rainfall_checker([70, 80, 65, 75], 70)
```

```python
# Problem 4: Duplicate Account Detection
def duplicate_accounts(usernames):
    duplicates = len(usernames) != len(set(usernames))
    print("Duplicate Accounts Found:", "Yes" if duplicates else "No")

duplicate_accounts(["user1", "user2", "user1"])
```

```python
# Problem 5: Appointment Eligibility Checker
def appointment_eligibility(age):
    status = "Eligible" if age >= 18 else "Not Eligible"
    print("Patient Age:", age)
    print("Eligibility Status:", status)

appointment_eligibility(21)
```

```python
# Problem 6: Premium Crop Price Filter
def premium_crops(prices):
    premium = [price for price in prices if price > 2000]
    print("Premium Crops:", premium)

premium_crops([1500, 2500, 3200, 1800])
```

```python
# Problem 7: Application Health Checker
def app_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    print("Error Count:", errors)
    print("System Status:", status)

app_health(7)
```

```python
# Problem 8: Daily Transaction Limit Checker
def transaction_check(amount):
    status = "Approved" if amount <= 50000 else "Rejected"
    print("Transaction Amount:", amount)
    print("Transaction Status:", status)

transaction_check(60000)
```

```python
# Problem 9: Attendance Eligibility System
def attendance_eligibility(attendance):
    present = attendance.count("P")
    percentage = (present / len(attendance)) * 100
    status = "Eligible" if percentage >= 75 else "Not Eligible"
    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)

attendance_eligibility(["P", "P", "A", "P", "P"])
```

```python
# Problem 10: Smart Electricity Bill Analyzer
def electricity_bill(units):
    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    print("Units Consumed:", units)
    print("Total Bill:", bill)
    print("Usage Level:", usage)

electricity_bill(250)
```
